```python
try:
    prices['date'] = pd.to_datetime(prices['date'], format='%Y-%m-%d')  # Specify format
except ValueError:
    # Attempt 2: Use 'mixed' format if dates are inconsistent
    prices['date'] = pd.to_datetime(prices['date'], format='mixed', errors='coerce')
latest_dates = prices.groupby('symbol')['date'].nlargest(5).reset_index()
latest_dates = pd.merge(latest_dates, prices, on=['symbol', 'date'], how='left')
latest_dates = latest_dates.drop('level_1', axis=1)
```

```python
latest_dates.head(10)
```

```python
latest_dates['date_rank'] = latest_dates.groupby('symbol')['date'].rank(ascending=True, method='first').astype(int)
pivoted_dates = latest_dates.pivot(index='symbol', columns='date_rank', values=['open', 'close', 'low', 'high'])
pivoted_dates.columns = ['_'.join(map(str, col)).strip() for col in pivoted_dates.columns]
pivoted_dates = pivoted_dates.reset_index()

print(pivoted_dates)
```

```python
pivoted_dates.columns
```

```python
pivoted_dates['open_media'] = (pivoted_dates['open_1'] + pivoted_dates['open_2'] +
                                pivoted_dates['open_3'] + pivoted_dates['open_4'] + pivoted_dates['open_5'] ) / 5
pivoted_dates.drop(['open_1', 'open_2', 'open_3', 'open_4', 'open_5'], axis=1, inplace=True)
```

```python
pivoted_dates['close_media'] = (pivoted_dates['close_1'] + pivoted_dates['close_2'] +
                                pivoted_dates['close_3'] + pivoted_dates['close_4'] + pivoted_dates['close_5'] ) / 5
pivoted_dates.drop(['close_1', 'close_2', 'close_3', 'close_4', 'close_5'], axis=1, inplace=True)
```

```python
pivoted_dates['low_media'] = (pivoted_dates['low_1'] + pivoted_dates['low_2'] +
                                pivoted_dates['low_3'] + pivoted_dates['low_4'] + pivoted_dates['low_5'] ) / 5
pivoted_dates.drop(['low_1', 'low_2', 'low_3', 'low_4', 'low_5'], axis=1, inplace=True)
```

```python
pivoted_dates['high_media'] = (pivoted_dates['high_1'] + pivoted_dates['high_2'] +
                                pivoted_dates['high_3'] + pivoted_dates['high_4'] + pivoted_dates['high_5'] ) / 5
pivoted_dates.drop(['high_1', 'high_2', 'high_3', 'high_4', 'high_5'], axis=1, inplace=True)
```
